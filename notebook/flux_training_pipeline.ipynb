```python
import os
import random
import torch
from diffusers import FluxPipeline

# ==============================
# CONFIG
# ==============================
MODEL_ID = "black-forest-labs/FLUX.1-dev"
LORA_DIR = "/teamspace/studios/this_studio/output/flux-mughal-lora-v1"
OUTPUT_DIR = "/teamspace/studios/this_studio/generated_images"

NUM_IMAGES = 250

os.makedirs(OUTPUT_DIR, exist_ok=True)

# ==============================
# LOAD PIPELINE
# ==============================
pipe = FluxPipeline.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.bfloat16
)

pipe.enable_model_cpu_offload()
pipe.enable_attention_slicing()

pipe.load_lora_weights(
    LORA_DIR,
    weight_name="flux-mughal-lora-v1.safetensors",
    adapter_name="mughal"
)

# ==============================
# PROMPT GENERATOR
# ==============================

import random

def generate_prompt():

    base = "MGHML, mughal miniature painting, opaque watercolor and gold on paper"

    constraint = "flat 2D perspective, clear figure separation, no overlapping bodies, no extra arms, no extra legs, correct anatomy, single head per figure, not photorealistic, no 3D rendering, no Western perspective"

    scene = random.choice([
        "court", "hunt", "portrait", "procession",
        "falconry", "garden", "battle", "scholar", "music"
    ])

    # ===== COURT =====
    if scene == "court":
        subject = "emperor seated on elevated throne"
        composition = "symmetrical arrangement with four courtiers standing separately on both sides"
        environment = "palace courtyard with arches and marble floor"
        detail = random.choice([
            "attendants holding flywhisks",
            "rich carpet with geometric patterns",
            "gold ornamentation on throne"
        ])

    # ===== HUNT =====
    elif scene == "hunt":
        subject = "emperor riding a horse hunting a deer"
        composition = "three figures placed separately with clear spacing"
        environment = "rocky landscape with sparse trees and flowing stream"
        detail = random.choice([
            "attendants following at distance",
            "dogs chasing deer",
            "dust trail behind horses"
        ])

    # ===== PORTRAIT =====
    elif scene == "portrait":
        subject = "single nobleman standing in profile holding a falcon"
        composition = "single central figure with no additional figures"
        environment = "flat dark green background"
        detail = random.choice([
            "pearl necklaces and jeweled turban",
            "gold patterned jama",
            "refined facial detailing"
        ])

    # ===== PROCESSION =====
    elif scene == "procession":
        subject = "large elephant carrying two nobles in ornate howdah"
        composition = "figures aligned horizontally with clear spacing"
        environment = "rolling hills with shrubs and distant riders"
        detail = random.choice([
            "pink and gold caparison",
            "attendants carrying ceremonial fans",
            "textile ornamentation"
        ])

    # ===== FALCONRY =====
    elif scene == "falconry":
        subject = "nobleman holding a falcon on raised arm"
        composition = "two figures with clear separation"
        environment = "garden with trees and water channel"
        detail = random.choice([
            "attendant standing behind",
            "rich garments with gold embroidery",
            "calm aristocratic posture"
        ])

    # ===== GARDEN =====
    elif scene == "garden":
        subject = "three noblewomen gathering flowers"
        composition = "three figures placed separately without overlap"
        environment = "lush garden with flowering trees and plants"
        detail = random.choice([
            "baskets of flowers",
            "birds in background",
            "fountain at center"
        ])

    # ===== BATTLE =====
    elif scene == "battle":
        subject = "two groups of cavalry facing each other"
        composition = "clearly separated horses and riders"
        environment = "open dusty battlefield with flags"
        detail = random.choice([
            "warriors with swords",
            "horses in motion",
            "layered formation"
        ])

    # ===== SCHOLAR =====
    elif scene == "scholar":
        subject = "single scholar seated writing manuscript"
        composition = "single figure with objects placed around"
        environment = "simple interior with books and scrolls"
        detail = random.choice([
            "ink pot and tools",
            "scrolls on floor",
            "quiet focused posture"
        ])

    # ===== MUSIC =====
    elif scene == "music":
        subject = "three musicians seated playing instruments"
        composition = "figures arranged with equal spacing"
        environment = "indoor pavilion with carpet and cushions"
        detail = random.choice([
            "court audience behind",
            "sitar and drums",
            "decorated interior"
        ])

    prompt = f"{base}, {subject}, {composition}, {environment}, {detail}, {constraint}"

    return prompt


# ==============================
# CLEAN FILE NAME
# ==============================
def clean_filename(text):
    return text.replace(",", "").replace(" ", "_")[:80]


# ==============================
# GENERATION LOOP
# ==============================
for i in range(NUM_IMAGES):

    prompt = generate_prompt()

    # LoRA variation
    lora_weight = random.choice([0.7, 0.8, 0.9])
    pipe.set_adapters("mughal", adapter_weights=lora_weight)

    generator = torch.manual_seed(i)

    image = pipe(
        prompt,
        height=1024,
        width=1024,
        guidance_scale=4.5,
        num_inference_steps=25,
        generator=generator
    ).images[0]

    filename = clean_filename(prompt)

    # Save image
    image.save(f"{OUTPUT_DIR}/{filename}_{i}.png")

    # Save prompt (for CLIP)
    with open(f"{OUTPUT_DIR}/{filename}_{i}.txt", "w") as f:
        f.write(prompt)

    if i % 20 == 0:
        print(f"Generated {i}/{NUM_IMAGES}")

print("Done ✅")
```

```python
import os
import torch
from diffusers import FluxPipeline

# =========================
# CONFIG
# =========================
MODEL_ID = "black-forest-labs/FLUX.1-dev"
LORA_PATH = "/teamspace/studios/this_studio/output/flux-mughal-lora-v1/flux-mughal-lora-v1.safetensors"
OUTPUT_DIR = "/teamspace/studios/this_studio/generated_images_up"

os.makedirs(OUTPUT_DIR, exist_ok=True)

# =========================
# MEMORY SETTINGS
# =========================
os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "expandable_segments:True"

# =========================
# LOAD PIPELINE
# =========================
pipe = FluxPipeline.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.float16
)

pipe.enable_sequential_cpu_offload()
pipe.enable_attention_slicing("max")
pipe.vae.enable_slicing()
pipe.vae.enable_tiling()

pipe.load_lora_weights(
    os.path.dirname(LORA_PATH),
    weight_name=os.path.basename(LORA_PATH),
    adapter_name="mughal"
)

# =========================
# YOUR 20 PROMPTS
# =========================
prompts = [

"MGHML, mughal miniature painting, opaque watercolor and gold on paper, emperor seated watching dancers performing in courtyard with two attendants behind, palace arches and carpet, flat 2D, clear spacing, no overlapping figures, correct anatomy",

"MGHML, mughal miniature painting, opaque watercolor and gold on paper, nobleman feeding deer in enclosed garden with one attendant behind, trees and flowers, calm scene, flat 2D perspective, no extra limbs",

"MGHML, mughal miniature painting, opaque watercolor and gold on paper, prince seated listening to storyteller with one attendant beside, indoor pavilion with cushions, flat 2D composition, clear figure separation",

"MGHML, mughal miniature painting, opaque watercolor and gold on paper, nobleman inspecting jewels placed on tray with one attendant behind, interior setting, flat 2D, fine detailing, no distortion",

"MGHML, mughal miniature painting, opaque watercolor and gold on paper, emperor observing wrestling match between two figures with attendants behind, open courtyard, flat 2D perspective, no overlapping bodies",

"MGHML, mughal miniature painting, opaque watercolor and gold on paper, noblewoman applying henna while seated with one attendant assisting, interior chamber, flat 2D composition, delicate detailing",

"MGHML, mughal miniature painting, opaque watercolor and gold on paper, royal cook preparing food near vessels with one noble observing, simple courtyard, flat 2D, clear spacing, no extra limbs",

"MGHML, mughal miniature painting, opaque watercolor and gold on paper, nobleman watering plants in garden with one attendant behind, flowering trees and pots, flat 2D composition, no overlapping figures",

"MGHML, mughal miniature painting, opaque watercolor and gold on paper, emperor seated watching elephant fight from distance with attendants beside, open ground, flat 2D perspective, clear separation",

"MGHML, mughal miniature painting, opaque watercolor and gold on paper, nobleman arranging flowers in vase with one attendant behind, indoor setting, flat 2D composition, no distortion",

"MGHML, mughal miniature painting, opaque watercolor and gold on paper, prince playing board game with one companion seated opposite, carpet and cushions, flat 2D perspective, no overlapping bodies",

"MGHML, mughal miniature painting, opaque watercolor and gold on paper, nobleman receiving gifts from servant with items placed on tray, courtyard with arches, flat 2D composition, clear spacing",

"MGHML, mughal miniature painting, opaque watercolor and gold on paper, emperor watching horse race from terrace with attendants behind, distant riders visible, flat 2D perspective, no distortion",

"MGHML, mughal miniature painting, opaque watercolor and gold on paper, nobleman reading letter while seated with one attendant behind, interior pavilion, flat 2D composition, no extra limbs",

"MGHML, mughal miniature painting, opaque watercolor and gold on paper, woman holding mirror while adjusting jewelry with one attendant behind, interior chamber, flat 2D, fine detailing",

"MGHML, mughal miniature painting, opaque watercolor and gold on paper, nobleman playing flute seated in garden with one listener nearby, trees and birds, flat 2D perspective, no overlapping figures",

"MGHML, mughal miniature painting, opaque watercolor and gold on paper, emperor inspecting soldiers standing in line with clear spacing, courtyard setting, flat 2D composition, no distortion",

"MGHML, mughal miniature painting, opaque watercolor and gold on paper, nobleman lighting lamp at evening with one attendant behind, terrace with dark sky, flat 2D perspective, no extra limbs",

"MGHML, mughal miniature painting, opaque watercolor and gold on paper, prince observing painting being made by artist with one attendant nearby, interior setting, flat 2D composition",

"MGHML, mughal miniature painting, opaque watercolor and gold on paper, nobleman seated watching birds flying near water with one attendant behind, garden setting, flat 2D, clear separation"

]

# =========================
# GENERATION LOOP
# =========================
existing_files = [f for f in os.listdir(OUTPUT_DIR) if f.endswith(".png")]
start_index = len(existing_files)
for i, prompt in enumerate(prompts):
    idx = start_index + i

    torch.cuda.empty_cache()

    pipe.set_adapters("mughal", adapter_weights=0.9)

    image = pipe(
        prompt,
        height=1024,
        width=1024,
        guidance_scale=4.5,
        num_inference_steps=20
    ).images[0]

    # Save image
    image.save(f"{OUTPUT_DIR}/img_{idx}.png")

    with open(f"{OUTPUT_DIR}/img_{idx}.txt", "w") as f:
        f.write(prompt)

    print(f"Generated image {i+1}/20")

print("Done ✅")
```
